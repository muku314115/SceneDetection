```
#ONLY FOR FLICKR
import re
import requests
from bs4 import BeautifulSoup
from PIL import Image, ExifTags
from io import BytesIO

#site = 'https://www.flickr.com/explore'
site = 'https://www.flickr.com/photos/141278164@N06/'
response = requests.get(site)

soup = BeautifulSoup(response.text, 'html.parser')

containers=soup.findAll('div',
             {'class': lambda x: x 
                       and 'photo' in x.split('-') and 'view' in x.split('-')
             }
            )
urls=[]

containers_list = str(containers).split('\n')

for container in containers_list:
    container = container.replace("<", " ")
    container = container.replace(">", " ")
    url = container.split()[-1]
    if "url" in url:
      urls.append('http://' + url[6:-2])


#print(img_tags)



for url in urls:
    response = requests.get(url)
    img = Image.open(BytesIO(response.content))
    info = img.info
    print('\n')
    try:
        exif = { ExifTags.TAGS[k]: v for k, v in img._getexif().items() if k in ExifTags.TAGS }
        print(exif)
    except:
      print("no exif")
    
    print('\n')

    for item in info:
        print(item, end=": ")
        print(info[item])
```
